```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("results.csv")

# Parse the JSON-style columns
df["hypervolumes"] = df["hypervolumes"].apply(json.loads)
df["times"] = df["times"].apply(json.loads)

# Plot settings
plt.style.use('seaborn-darkgrid')
colors = plt.cm.tab10.colors

# Plot hypervolume progression
plt.figure(figsize=(10, 6))
for i, row in df.iterrows():
    iterations = list(range(len(row["hypervolumes"])))
    plt.plot(iterations, row["hypervolumes"], label=f"{row['acq_func']} (batch={row['batch_size']}, iter={row['n_iter']})", color=colors[i % len(colors)])

plt.xlabel("Iteration")
plt.ylabel("Hypervolume")
plt.title("Hypervolume Progression Across Experiments")
plt.legend()
plt.tight_layout()
plt.savefig("hypervolume_progression.png")
plt.show()

# Plot cumulative time vs. iteration
plt.figure(figsize=(10, 6))
for i, row in df.iterrows():
    iterations = list(range(len(row["times"])))
    cumulative_times = pd.Series(row["times"]).cumsum()
    plt.plot(iterations, cumulative_times, label=f"{row['acq_func']} (batch={row['batch_size']}, iter={row['n_iter']})", color=colors[i % len(colors)])

plt.xlabel("Iteration")
plt.ylabel("Cumulative Time (s)")
plt.title("Cumulative Time vs. Iteration")
plt.legend()
plt.tight_layout()
plt.savefig("cumulative_time.png")
plt.show()
```
